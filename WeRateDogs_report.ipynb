## The Renowned WeRateDogs rating system

One of the things that makes the WeRateDogs page what it is is the tendency to give dog images ratings (normally out of 10) where the numerator is greater than the denominator. For example, a common rating given is 13/10. From a data analysis standpoint, looking at scores that are greater than 100% is confusing, but from a human standpoint, it's very cute and contributes to the originality of this page. 

When I began looking at the scores, I assumed this was standard practice from the page's beginnings. However, the data shows otherwise. The graph below shows the proportion of hallmark WeRateDogs ratings ( numerator > denominator ) and the proportion of "normal ratings" <br>( numerator < denominator ) over the time frame the data spans, in months. 

In the beginning, the WeRateDogs style ratings only represented about 20% of the total ratings, but by November 2016, almost *all* of the ratings have the denominator greater than the numerator, and the trend doesn't appear to be slowing.

From the data, it seems the page's typical rating system started off as a few over-the-top ratings for exceptional dogs here and there, but as the page and the rating system gained traction, the average numerator rose above 10 never to go back down. 

## Dog Stages in WeRateDogs
Another salient aspect of the WeRateDogs Twitter page is its use of dog "stages" such as pupper, puppo, doggo and floofer to describe the dogs it rates. These stages correspond to a certain point in the life of a dog or some aspect of its appearance, as is the case of floofer, which is a very furry dog. The stages even have their own entries in the WeRateDogs "Dogtionary", so it's safe to say they are part of what makes this page so unique. 

But have they been around since the beginning? Ordering the over two thousand WeRateDogs tweets between November 2015 and August 2017 by date, I found that no, they haven't. The pupper appeared first in December 2015 and has appeared most since. Followed by Floofer in January of 2016, doggo in April of 2016, and puppo in June of 2016. 

Perhaps since the pupper has been around longest, it has significantly more representation than the other dog stages. Just enough to make a look into its evolution over time interesting. The graphs below juxtapose the proportion of WeRateDogs pupper rating tweets each month and the median like count each month for pupper and not-puppers.


The first thing to notice is that as time passes, the proportion of tweets that are puppers diminishes. However, the median like count for puppers starts to rise faster than the median like count for all other tweets. It seems the WeRateDogs page began to use puppers more and more sparingly, but to greater effect. 

The trend is broken in June 2017, but comes back the following month. It would be interesting to see further into the history of the page to see if the trend continues.


## Most Predicted Breed Across months
Some of the data I gathered for this project were image predictions made by a neural network. The network tries to predict what dog breed is in each WeRateDogs tweet, and provides a confidence level for each prediction.

I thought it would be interesting to see if a predicted breed was predicted more than any other consistently across months, a sort of dog breed of the month in terms of predictions. 

Since these are only *predictions* and may or may not be true, I tried to be as accurate as possible while still retaining a good amount of tweets to work with. So I analyzed only the predictions of tweets with a confidence level greater than 0.80. 

The graph below shows the breeds that were predicted most in at least one month, as well as how many months they were predicted most. 

The golden retriever is the most predicted breed 7 months out of the 22 analyzed. This seems very consistent, even when you consider the limitations set by confidence level and the fact that these are predictions. However me mustn't forget these limitations. For the moment, we can say, very tentatively, that the golden retriever is the most predicted breed across months. 



# Conclusions

To summarize, WeRateDogs is a dynamic page whose style has evolved as its popularity has grown. This evolution has led to the development of a signature rating system and a lot of fun canine terms. 

In spite of how the page has changed, one thing that appears to have stayed somewhat constant is the most tweeted breed per month, the golden retriever. But given the limitations of the analysis, this should be taken with a grain of salt.
